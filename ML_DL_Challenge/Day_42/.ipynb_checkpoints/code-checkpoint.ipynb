```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np 

X,y = make_moons(n_samples=1000, noise=0.05)
dbscan_model = DBSCAN(eps=0.05, min_samples=5)
dbscan_model.fit(X)
#print(dbscan_model.labels_)
print(len(dbscan_model.components_))
#print(dbscan_model.core_sample_indices_)
print(dbscan_model.components_)
print(np.unique(dbscan_model.labels_))
dbscan_model2 = DBSCAN(eps=0.2, min_samples=5)
dbscan_model2.fit(X)

def plot_dbscan(model,X,showY=True,showX=True):
    core = np.zeros_like(model.labels_, dtype=bool)
    core[model.core_sample_indices_] = True
    outliers = model.labels_ == -1
    noncore = ~(core | outliers)
    core_comp =  model.components_
    outlier_comp = X[outliers]
    noncore_comp = X[noncore]
    plt.scatter(core_comp[:,0],core_comp[:,1],marker="o", s=200, c=model.labels_[core],cmap="Spectral")
    plt.scatter(core_comp[:, 0], core_comp[:, 1], marker='*', s=20, c=model.labels_[core])
    plt.scatter(outlier_comp[:, 0], outlier_comp[:, 1],c="r",marker="x",label="Outliers")
    plt.scatter(noncore_comp[:, 0], noncore_comp[:, 1],c=model.labels_[noncore],marker=".",s=10,label="Non Core points")
    plt.title (f"eps:{model.eps}, min_samples:{model.min_samples}", fontsize=12)
    plt.xlabel("X")
    plt.ylabel("y")
    if (showY == False):
        plt.tick_params(labelleft=False)
        plt.ylabel("")
    

plt.figure(figsize=(10,4))
plt.subplot(121)
plot_dbscan(dbscan_model,X)
plt.subplot(122)
plot_dbscan(dbscan_model2,X,showY=False)
plt.suptitle("DBSCAN clustering using two different neighborhood radiuses", fontsize=15)
plt.tight_layout()
```

```python
# Prediction in DBSCAN
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier (n_neighbors=50)
knn_model.fit(dbscan_model2.components_,dbscan_model2.labels_[dbscan_model2.core_sample_indices_])
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
print(knn_model.predict(X_new))
print(knn_model.predict_proba(X_new))
```

```python
# Other Clustering Algorithm
# Spectral

from sklearn.cluster import SpectralClustering

spectral_model1 = SpectralClustering(n_clusters=2, gamma=100, random_state=42)
spectral_model1.fit(X)

spectral_model2 = SpectralClustering(n_clusters=2, gamma=1, random_state=42)
spectral_model2.fit(X)

print(np.percentile(spectral_model1.affinity_matrix_,95))
print(np.percentile(spectral_model2.affinity_matrix_,95))
```

```python
def plot_spectral(model,X,showY=True):
    plt.scatter(X[:,0],X[:,1],marker="o",c="gray",s=500,alpha=0.1)
    plt.scatter(X[:,0],X[:,1],marker="o",c="white",s=30)
    plt.scatter(X[:,0],X[:,1],marker=".",s=10,c=model.labels_, cmap="Paired")
    plt.xlabel("X")
    plt.ylabel("y")
    if showY == False:
        plt.tick_params(labelleft=False)
        plt.ylabel("")
    plt.title (f"Gamma: {model.gamma}", fontsize=12)

plt.figure(figsize=(10,4))
plt.subplot(121)
plot_spectral(spectral_model1,X)
plt.subplot(122)
plot_spectral(spectral_model2,X,showY=False)
plt.suptitle("Cluster Modeling using two different Kernel coefficient", fontsize=15)
plt.tight_layout()
```

```python
# Agglomerative Algorithm

from sklearn.cluster import AgglomerativeClustering

agg_model = AgglomerativeClustering(n_clusters=2,linkage="complete")
agg_model.fit(X)
print(agg_model.labels_)
```
