```python
import tensorflow as tf
from d2l import tensorflow as d2l
```

```python
# Concise Implementation of Multilayer Perceptrons
model = tf.keras.Sequential([
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation="relu"),      # First hidden layer
        tf.keras.layers.Dense(10)                           # Second hidden layer
])
batchsize=256
lr=0.1
num_epochs=10
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optimizer = tf.keras.optimizers.SGD(learning_rate=lr)
train,test=d2l.load_data_fashion_mnist(batch_size=batchsize)
```

```python
d2l.train_ch3(model,train,test,loss,num_epochs,optimizer)
```
