```python
import tensorflow as tf
from tensorflow import keras

# Simple Perceptron
NB_CLASSES=10
RESHAPED=784
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(NB_CLASSES,
                             input_shape=(RESHAPED,),
                             kernel_initializer="zeros",  # Updating weights, other values are random_uniform and random_normal
                             name="dense",
                             activation = "softmax"
                             ))
```

```python
# Activation function
import numpy as np
import matplotlib.pyplot as plt

def threshold(x):
  cond = tf.less(x,tf.zeros(tf.shape(x), dtype=x.dtype))
  out = tf.where(cond,tf.zeros(tf.shape(x)),tf.ones(tf.shape(x)))
  return out

plt.figure(figsize=(12,8))
def plot_activation(x,y,title,plot):
  plt.subplot(plot)
  plt.plot(x,y)
  plt.title(title,fontsize=12)
  if (plot == 341 or  plot == 345):
    plt.ylabel("Output of Neuron")
  if (plot > 344):
    plt.xlabel("Activity of Neuron")
  
x = np.linspace(-10,10,200)
y = threshold(x)
plot_activation(x,y,"Threshold",341)
plot_activation(x,tf.sigmoid(x),"Sigmoid",342)
plot_activation(x,tf.tanh(x),"Hyperbolic Tangent",343)
plot_activation(x,tf.nn.relu(x),"Relu",344)
plot_activation(x,tf.nn.leaky_relu(x),"Leaky Relu",345)
plot_activation(x,tf.nn.elu(x),"Elu",346)
plot_activation(x,tf.nn.softmax(x),"Softmax",347)
plot_activation(x,tf.keras.activations.softplus(x),"Softplus",348)
plt.tight_layout()
```
